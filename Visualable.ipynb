```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
```

```python
image = cv.imread(r'Images/testimg2.jpg')
hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)

blurimg = cv.imread(r'Images/blurimg2.jpg')
blurhsv = cv.cvtColor(blurimg, cv.COLOR_BGR2HSV)
```

```python
HueUnique = np.unique(hsv[:,:,0])
HueUnique = HueUnique.astype('int32')
HueUnique = HueUnique.reshape(HueUnique.size, 1)
hue_values = np.concatenate((HueUnique, np.ones((HueUnique.size, 1), dtype='int32')), axis=1)

for u in range(HueUnique.size):
    hue_values[u, 1] = np.sum(hsv[:,:,0]==hue_values[u, 0])
```

```python
## convert your array into a dataframe
df = pd.DataFrame (hue_values)

## save to xlsx file

filepath = 'my_excel_file.xlsx'

df.to_excel(filepath, index=False)
```

```python
#plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
x = hue_values[:, 0]
y = hue_values[:, 1]
ax.plot(x, y)
```
